```python
import psycopg2 as pg
import pandas.io.sql as psql
import pandas as pd
import numpy as np
import gensim
import re
from tqdm import tqdm, tqdm_notebook
import logging
logging.basicConfig(format="%(levelname)s - %(asctime)s: %(message)s", datefmt= '%H:%M:%S', level=logging.INFO)
from time import time 
import spacy
```

```python
# Input all imaging reports
data = pd.read_csv('mimic_rad.csv',header=None)
#data
```

```python
# data cleaning
t = time()

brief_cleaning = [row.lower() for row in data[0]]

print('Time to clean up everything: {} mins'.format(round((time() - t) / 60, 2)))
```

```python
#check data
for doc in brief_cleaning: 
    print(doc)
    break
```

```python
# data cleaning and create datasets
corpus = [doc.split() for doc in tqdm_notebook(brief_cleaning)]
```

```python
len(corpus)
```

```python
from gensim.models.phrases import Phrases, Phraser
```

```python
# Create phrases
phrases = Phrases(corpus, min_count=30, progress_per=10000)
```

```python
bigram = Phraser(phrases)
corpus = bigram[corpus]
```

```python
import multiprocessing
cores = multiprocessing.cpu_count()
cores
```

```python
# Create w2v model using gensim
model = gensim.models.Word2Vec(
        corpus,
        size=300,
        window=10,
        min_count=10,
        workers=cores-1,
        negative=15)

# with hidden layer of 300 dimensions, window of 10 neighboring words, and minimun word counts of 10. 
```

```python
model.init_sims(replace=True)
```

```python
#save w2v
from gensim.test.utils import get_tmpfile
from gensim.models import KeyedVectors
word_vectors = model.wv
fname = get_tmpfile('mimic3_rad_sim.kv')
word_vectors.save(fname)
word_vectors = KeyedVectors.load(fname, mmap='r')
```

```python
#Test w2v
word_vectors.most_similar(positive='pneumonia',topn=20)
```
